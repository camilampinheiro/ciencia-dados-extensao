## 1. Importando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Carregando o dataset

```python
df = pd.read_csv("datasets/brazilian_cities.csv")
```

```python
df.head()
```

Para facilitar a análise, vamos selecionar algumas colunas importantes.

```python
df = df[[
    "CITY",
    "STATE",
    "CAPITAL",
    "ESTIMATED_POP",
    "AREA",
    "GDP",
    "GDP_CAPITA",
    "IDHM"
]]
```

Também vamos garantir que algumas colunas estejam no formato numérico.

```python
df["ESTIMATED_POP"] = pd.to_numeric(df["ESTIMATED_POP"], errors="coerce")
df["GDP"] = pd.to_numeric(df["GDP"], errors="coerce")
df["GDP_CAPITA"] = pd.to_numeric(df["GDP_CAPITA"], errors="coerce")
df["AREA"] = pd.to_numeric(df["AREA"], errors="coerce")
```

## 3. Introdução ao Matplotlib

O **Matplotlib** é a biblioteca base de visualização em Python.

Com ela, podemos criar gráficos como:

- linhas;
- barras;
- dispersão (scatter);
- histogramas.

## Gráfico de linha

```python
df_pop = df.sort_values("ESTIMATED_POP").head(10)

plt.plot(df_pop["CITY"], df_pop["ESTIMATED_POP"])
plt.xticks(rotation=45)
plt.title("População de algumas cidades")
plt.xlabel("Cidade")
plt.ylabel("População")
plt.show()
```

O gráfico de linha é útil para mostrar evolução ou comparação simples entre valores.

## Gráfico de barras

```python
df_pop = df.sort_values("ESTIMATED_POP", ascending=False).head(10)

plt.bar(df_pop["CITY"], df_pop["ESTIMATED_POP"])
plt.xticks(rotation=45)
plt.title("Top 10 cidades por população")
plt.show()
```

O gráfico de barras é ideal para comparar categorias.

## Gráfico de dispersão (scatter)

```python
plt.scatter(df["GDP_CAPITA"], df["IDHM"])
plt.title("Relação entre GDP per capita e IDHM")
plt.xlabel("GDP per capita")
plt.ylabel("IDHM")
plt.show()
```

O gráfico de dispersão é utilizado para observar relação entre duas variáveis.

## Histograma

```python
plt.hist(df["IDHM"].dropna(), bins=20)
plt.title("Distribuição do IDHM")
plt.xlabel("IDHM")
plt.ylabel("Frequência")
plt.show()
```

O histograma mostra como os valores estão distribuídos.

## 4. Introdução ao Seaborn

O **Seaborn** é uma biblioteca baseada no Matplotlib, mas com gráficos mais bonitos e fáceis de usar.

## Gráfico de barras com Seaborn

```python
df_pop = df.sort_values("ESTIMATED_POP", ascending=False).head(10)

sns.barplot(x="CITY", y="ESTIMATED_POP", data=df_pop)
plt.xticks(rotation=45)
plt.title("Top 10 cidades por população")
plt.show()
```

O Seaborn facilita a criação de gráficos mais organizados e visualmente agradáveis.

## Boxplot

```python
sns.boxplot(x=df["IDHM"])
plt.title("Boxplot do IDHM")
plt.show()
```

O boxplot ajuda a identificar:

- mediana;
- quartis;
- valores extremos (outliers).

## Heatmap (correlação)

```python
corr = df[["ESTIMATED_POP", "GDP", "GDP_CAPITA", "IDHM"]].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlação entre variáveis")
plt.show()
```

O heatmap mostra a relação entre variáveis numéricas.

Valores próximos de 1 indicam forte correlação positiva.

## 5. Interpretação dos gráficos

A visualização não serve apenas para “mostrar gráficos”, mas para **gerar insights**.

Exemplos:

- cidades com maior população tendem a ter maior PIB;
- pode existir relação entre GDP per capita e IDHM;
- a distribuição do IDHM pode indicar desigualdade entre municípios.
