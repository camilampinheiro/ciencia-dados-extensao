## 1. Por que visualizar dados?

Imagine receber uma tabela com 5.570 linhas - uma para cada município brasileiro. Como você identificaria padrões, outliers ou tendências apenas olhando números?

É por isso que existe a **visualização de dados**: transformar números em imagens que o cérebro humano consegue interpretar rapidamente.

### O que um bom gráfico deve fazer:
- Responder uma pergunta clara
- Ser fácil de entender em poucos segundos
- Usar o tipo de gráfico certo para a informação certa
- Ter título, eixos nomeados e legenda quando necessário

### O que um gráfico ruim faz:
- Confundir mais do que explicar
- Usar o tipo errado (ex: pizza com 20 fatias)
- Não ter contexto (eixos sem nome, sem título)
- Usar cores sem propósito


## 2. Tipos de gráficos e quando usar cada um

Antes de escrever qualquer código, você precisa fazer uma pergunta:

> **"O que eu quero mostrar com esse gráfico?"**

| Objetivo | Tipo de Gráfico | Exemplo de pergunta |
|---|---|---|
| **Comparar** valores entre grupos | Barras | Quais estados têm maior PIB? |
| **Distribuição** de uma variável | Histograma | Como o IDHM está distribuído? |
| **Distribuição** entre grupos | Boxplot | O IDHM varia entre cidades urbanas e rurais? |
| **Relação** entre duas variáveis | Scatterplot | Renda alta implica IDHM alto? |
| **Evolução** ao longo do tempo | Linha | O IDHM melhorou entre 1991 e 2010? |
| **Composição** de um todo | Pizza / Barras empilhadas | Qual setor domina a economia? |
| **Correlação** entre muitas variáveis | Heatmap | Quais variáveis se relacionam? |

Vamos praticar cada um desses tipos.

## 4. Matplotlib: a base de tudo

O **Matplotlib** é a biblioteca mais fundamental de visualização em Python. Toda visualização em Python, de alguma forma, passa por ela - inclusive o Seaborn é construído sobre o Matplotlib.

### Estrutura básica de um gráfico Matplotlib

```python
plt.figure(figsize=(largura, altura))  # 1. Cria a figura
plt.plot(x, y)                         # 2. Plota os dados
plt.title("Título")                    # 3. Define o título
plt.xlabel("Eixo X")                   # 4. Nome do eixo X
plt.ylabel("Eixo Y")                   # 5. Nome do eixo Y
plt.tight_layout()                     # 6. Ajusta os espaços
plt.show()                             # 7. Exibe o gráfico
```

Vamos ver isso na prática com exemplos simples antes de aplicar no dataset real.


### 4.1 Gráfico de linha - evolução ao longo do tempo

O gráfico de linha é perfeito para mostrar como algo **muda ao longo do tempo**.

Primeiro, um exemplo simples para entender a sintaxe:


````python
# ---- EXEMPLO DIDÁTICO: Gráfico de linha ----
# Dados fictícios: temperatura ao longo de uma semana
dias = ["Seg", "Ter", "Qua", "Qui", "Sex", "Sab", "Dom"]
temperatura = [28, 30, 27, 32, 35, 31, 29]

# Criando o gráfico passo a passo
plt.figure(figsize=(10, 5))          # Passo 1: define o tamanho

plt.plot(                            # Passo 2: plota a linha
    dias,                            # valores do eixo X
    temperatura,                     # valores do eixo Y
    color="steelblue",             # cor da linha
    linewidth=2,                     # espessura da linha
    marker="o",                    # símbolo nos pontos
    markersize=8                     # tamanho dos pontos
)

plt.title("Temperatura ao longo da semana", fontsize=14)  # Passo 3: título
plt.xlabel("Dia da semana")                               # Passo 4: eixo X
plt.ylabel("Temperatura (graus C)")                       # Passo 5: eixo Y
plt.tight_layout()                                        # Passo 6: ajusta
plt.show()                                                # Passo 7: exibe

# INTERPRETAÇÃO: O gráfico mostra que sexta-feira foi o dia mais quente.
# A linha conecta os pontos, deixando clara a tendência ao longo dos dias.
````

### 4.2 Gráfico de barras - comparação entre categorias

O gráfico de barras é o mais comum para **comparar valores entre grupos**.


````python
# ---- EXEMPLO DIDÁTICO: Gráfico de barras ----
# Dados fictícios: vendas por produto
produtos = ["Produto A", "Produto B", "Produto C", "Produto D"]
vendas = [450, 320, 580, 200]

plt.figure(figsize=(10, 5))

# barh = barras horizontais (mais legíveis quando os nomes são longos)
plt.barh(
    produtos,             # categorias no eixo Y
    vendas,               # valores no eixo X
    color="steelblue",   # cor das barras
    edgecolor="white"    # borda branca entre as barras
)

plt.title("Vendas por produto", fontsize=14)
plt.xlabel("Quantidade vendida")
plt.tight_layout()
plt.show()

# INTERPRETAÇÃO: O Produto C teve o maior volume de vendas (580 unidades),
# enquanto o Produto D teve o menor (200).
````

````python
# ---- EXEMPLO DIDÁTICO: Histograma ----
# Gerando dados fictícios: notas de 200 alunos
np.random.seed(42)
notas = np.random.normal(loc=7, scale=1.5, size=200)  # média 7, desvio 1.5
notas = np.clip(notas, 0, 10)  # limita entre 0 e 10

plt.figure(figsize=(10, 5))

plt.hist(
    notas,          # os dados
    bins=20,        # número de intervalos
    color="teal",  # cor das barras
    edgecolor="white",  # separação entre barras
    alpha=0.8       # transparência
)

# Linha vertical na média
plt.axvline(notas.mean(), color="red", linestyle="--", linewidth=2, label=f"Média: {notas.mean():.1f}")
plt.legend()

plt.title("Distribuição das notas dos alunos", fontsize=14)
plt.xlabel("Nota")
plt.ylabel("Número de alunos")
plt.tight_layout()
plt.show()

# INTERPRETAÇÃO: A maioria das notas está concentrada entre 6 e 8.
# A distribuição tem formato de sino (distribuição normal), com a média próxima de 7.
````

## 5. Seaborn: gráficos mais bonitos com menos código

O **Seaborn** é construído sobre o Matplotlib, mas oferece funções de alto nível que tornam a criação de gráficos mais simples e visualmente agradável.

### Diferença prática:
- **Matplotlib**: mais controle, mais código
- **Seaborn**: menos código, mais bonito por padrão

Na prática, usamos os dois juntos: Seaborn para criar o gráfico, Matplotlib para ajustar títulos e eixos.


### 5.1 Boxplot - distribuição por grupos

O boxplot (ou diagrama de caixa) mostra:
- **Linha central**: mediana (valor do meio)
- **Caixa**: onde estão 50% dos dados (do 1º ao 3º quartil)
- **Bigodes**: faixa de valores esperados
- **Pontos isolados**: possíveis outliers


````python
# ---- EXEMPLO DIDÁTICO: Boxplot ----
# Dados fictícios: salários por departamento
df_exemplo = pd.DataFrame({
    "departamento": ["TI"]*5 + ["RH"]*5 + ["Vendas"]*5,
    "salario": [5000, 6000, 5500, 7000, 4500,
               3500, 3800, 4000, 3200, 3600,
               4000, 6500, 5000, 8000, 3800]
})

plt.figure(figsize=(10, 5))

sns.boxplot(
    data=df_exemplo,
    x="departamento",    # eixo X: categorias
    y="salario",         # eixo Y: valores
    palette="Set2"       # paleta de cores
)

plt.title("Distribuição de salários por departamento", fontsize=14)
plt.xlabel("Departamento")
plt.ylabel("Salário (R$)")
plt.tight_layout()
plt.show()

# INTERPRETAÇÃO:
# - O departamento de TI tem maior mediana salarial
# - Vendas tem maior variação (caixa mais longa)
# - O ponto em Vendas acima de 8000 é um outlier (valor atípico)
````

### 5.2 Scatterplot - relação entre duas variáveis

O scatterplot (ou gráfico de dispersão) mostra a relação entre duas variáveis numéricas. Cada ponto representa uma observação.


````python
# ---- EXEMPLO DIDÁTICO: Scatterplot ----
# Dados fictícios: horas de estudo vs nota
np.random.seed(0)
horas = np.random.uniform(1, 10, 50)
nota = horas * 0.8 + np.random.normal(0, 1, 50)
nota = np.clip(nota, 0, 10)

df_estudo = pd.DataFrame({"horas_estudo": horas, "nota": nota})

plt.figure(figsize=(10, 5))

sns.scatterplot(
    data=df_estudo,
    x="horas_estudo",  # eixo X
    y="nota",          # eixo Y
    color="steelblue",
    alpha=0.7,          # transparência dos pontos (evita sobreposição)
    s=80               # tamanho dos pontos
)

# Linha de tendência
sns.regplot(
    data=df_estudo,
    x="horas_estudo",
    y="nota",
    scatter=False,        # não repete os pontos
    color="red",
    line_kws={"linewidth": 2, "linestyle": "--"}
)

plt.title("Relação entre horas de estudo e nota", fontsize=14)
plt.xlabel("Horas de estudo")
plt.ylabel("Nota")
plt.tight_layout()
plt.show()

# INTERPRETAÇÃO: Existe uma tendência positiva clara.
# A linha vermelha mostra essa tendência geral.
# Quem estuda mais tende a tirar notas maiores.
````

## 6. Aplicando no dataset real: municípios brasileiros

Agora vamos aplicar tudo o que aprendemos no dataset de municípios brasileiros.

Cada gráfico abaixo responde a uma pergunta real sobre os dados.


````python
# Preparação dos dados para os exemplos
if df is not None:
    colunas_uteis = [col for col in [
        "CITY", "STATE", "GDP", "GDP_CAPITA", "IDHM",
        "RURAL_URBAN", "GVA_AGROPEC", "GVA_INDUSTRY",
        "GVA_SERVICES", "GVA_PUBLIC", "GVA_TOTAL",
        "ESTIMATED_POP", "COMP_TOT", "HOTELS", "CAPITAL"
    ] if col in df.columns]

    df_vis = df[colunas_uteis].copy()
    df_vis = df_vis.dropna(subset=[c for c in ["IDHM", "GDP_CAPITA"] if c in df_vis.columns])

    print(f"Dataset preparado: {df_vis.shape[0]} municípios x {df_vis.shape[1]} colunas")
    print(f"Colunas disponíveis: {df_vis.columns.tolist()}")
else:
    df_vis = None
    print("Dataset não carregado. Execute a célula de carregamento primeiro.")
````

### 6.1 Pergunta: Quais são os estados com maior PIB per capita médio?

**Tipo de gráfico:** Barras (comparação entre grupos)


````python
if df_vis is not None and "STATE" in df_vis.columns and "GDP_CAPITA" in df_vis.columns:

    # Calcula o PIB per capita médio por estado e pega os 10 maiores
    gdp_por_estado = (
        df_vis.groupby("STATE")["GDP_CAPITA"]  # agrupa por estado
        .mean()                                   # calcula a média
        .sort_values(ascending=False)             # ordena do maior para o menor
        .head(10)                                 # pega apenas os 10 primeiros
        .reset_index()                            # transforma em DataFrame
    )
    gdp_por_estado.columns = ["Estado", "PIB_per_capita_medio"]

    plt.figure(figsize=(12, 6))

    sns.barplot(
        data=gdp_por_estado,
        x="Estado",
        y="PIB_per_capita_medio",
        palette="Blues_d"   # gradiente de azul (mais escuro = maior valor)
    )

    plt.title("Top 10 estados com maior PIB per capita médio", fontsize=14)
    plt.xlabel("Estado")
    plt.ylabel("PIB per capita médio (R$)")
    plt.tight_layout()
    plt.show()

    print("\n Interpretação:")
    print(f"O estado com maior PIB per capita médio é {gdp_por_estado.iloc[0]['Estado']},")
    print(f"com média de R$ {gdp_por_estado.iloc[0]['PIB_per_capita_medio']:,.0f}.")
````

### 6.2 Pergunta: Como o IDHM está distribuído nos municípios?

**Tipo de gráfico:** Histograma com KDE (distribuição)


````python
if df_vis is not None and "IDHM" in df_vis.columns:

    plt.figure(figsize=(12, 6))

    # histplot com kde=True desenha o histograma E a curva de densidade
    sns.histplot(
        df_vis["IDHM"].dropna(),   # os dados (removendo nulos)
        bins=30,                     # 30 intervalos
        kde=True,                    # linha de densidade por cima
        color="teal",
        edgecolor="white"
    )

    # Linhas verticais na média e mediana
    media_idhm = df_vis["IDHM"].mean()
    mediana_idhm = df_vis["IDHM"].median()

    plt.axvline(media_idhm, color="red", linestyle="--", linewidth=2, label=f"Média: {media_idhm:.3f}")
    plt.axvline(mediana_idhm, color="orange", linestyle="--", linewidth=2, label=f"Mediana: {mediana_idhm:.3f}")

    plt.title("Distribuição do IDHM nos municípios brasileiros", fontsize=14)
    plt.xlabel("IDHM")
    plt.ylabel("Número de municípios")
    plt.legend()
    plt.tight_layout()
    plt.show()

    print("\n Interpretação:")
    print(f"A maioria dos municípios tem IDHM entre 0.5 e 0.7.")
    print(f"Média: {media_idhm:.3f} | Mediana: {mediana_idhm:.3f}")
````

### 6.3 Pergunta: O IDHM varia entre municípios urbanos e rurais?

**Tipo de gráfico:** Boxplot (distribuição por grupos)


````python
if df_vis is not None and "RURAL_URBAN" in df_vis.columns and "IDHM" in df_vis.columns:

    # Limpa a coluna RURAL_URBAN
    df_box = df_vis[df_vis["RURAL_URBAN"].notna()].copy()
    df_box = df_box[~df_box["RURAL_URBAN"].isin(["0", "Sem classificação"]) ]

    plt.figure(figsize=(12, 6))

    sns.boxplot(
        data=df_box,
        x="RURAL_URBAN",
        y="IDHM",
        palette="Set2",
        order=sorted(df_box["RURAL_URBAN"].unique())
    )

    plt.title("Distribuição do IDHM por tipologia rural/urbana", fontsize=14)
    plt.xlabel("Tipologia")
    plt.ylabel("IDHM")
    plt.xticks(rotation=30, ha="right")
    plt.tight_layout()
    plt.show()

    print("\n Interpretação:")
    print("Municípios urbanos tendem a ter IDHM mais alto e mais concentrado.")
    print("Municípios rurais apresentam maior variação e medianas mais baixas.")
````

### 6.4 Pergunta: PIB per capita alto implica IDHM alto?

**Tipo de gráfico:** Scatterplot (relação entre variáveis)


````python
if df_vis is not None and "GDP_CAPITA" in df_vis.columns and "IDHM" in df_vis.columns:

    # Filtra outliers extremos para melhorar a visualização (mantém 99% dos dados)
    p99 = df_vis["GDP_CAPITA"].quantile(0.99)
    df_scatter = df_vis[df_vis["GDP_CAPITA"] <= p99].copy()

    plt.figure(figsize=(12, 6))

    sns.scatterplot(
        data=df_scatter,
        x="GDP_CAPITA",
        y="IDHM",
        hue="RURAL_URBAN" if "RURAL_URBAN" in df_scatter.columns else None,
        alpha=0.5,
        palette="Set2",
        s=30
    )

    plt.title("Relação entre PIB per capita e IDHM", fontsize=14)
    plt.xlabel("PIB per capita (R$)")
    plt.ylabel("IDHM")
    plt.legend(title="Tipologia", bbox_to_anchor=(1.05, 1))
    plt.tight_layout()
    plt.show()

    # Calculando correlação
    corr = df_scatter[["GDP_CAPITA", "IDHM"]].corr().iloc[0,1]
    print(f"\n Correlação entre PIB per capita e IDHM: {corr:.3f}")
    print("Interpretação: Correlação positiva - municípios mais ricos tendem a ter maior IDHM.")
````

### 6.5 Mapa de correlação - relações entre múltiplas variáveis

**Tipo de gráfico:** Heatmap (correlações)


````python
if df_vis is not None:

    # Seleciona variáveis numéricas relevantes
    colunas_corr = [c for c in [
        "GDP", "GDP_CAPITA", "IDHM", "GVA_TOTAL",
        "ESTIMATED_POP", "COMP_TOT", "HOTELS"
    ] if c in df_vis.columns]

    # Calcula a matriz de correlação
    corr_matrix = df_vis[colunas_corr].corr(numeric_only=True)

    plt.figure(figsize=(10, 8))

    sns.heatmap(
        corr_matrix,
        annot=True,          # mostra os valores dentro das células
        fmt=".2f",           # 2 casas decimais
        cmap="coolwarm",     # vermelho = positivo, azul = negativo
        center=0,            # centraliza a escala de cor no zero
        linewidths=0.5,      # linhas entre as células
        square=True          # células quadradas
    )

    plt.title("Mapa de correlação entre variáveis econômicas", fontsize=14)
    plt.tight_layout()
    plt.show()

    print("\n Como ler este gráfico:")
    print("Vermelho forte = correlação positiva alta (quando um sobe, o outro também sobe)")
    print("Azul forte = correlação negativa alta (quando um sobe, o outro desce)")
    print("Próximo de zero = pouca relação linear")
````

## 7. Boas práticas de visualização

Agora que você já criou vários gráficos, veja o que separa uma visualização boa de uma excelente:

### Faça sempre:
- Use **títulos descritivos** (não "Gráfico 1", mas "Distribuição do IDHM por estado")
- **Nomeie os eixos** com o que representam e a unidade (ex: "PIB per capita (R$)")
- Use **cores com propósito** - cada cor deve representar uma categoria ou um valor
- **Reduza a poluição visual** - retire o que não agrega informação
- **Inclua a interpretação** - um gráfico sem explicação não comunica nada

### Evite:
- Gráficos de pizza com mais de 5 categorias
- Escalas que distorcem a percepção (começar o eixo Y em um valor arbitrário)
- Excesso de cores sem significado
- Títulos vagos como "Dados" ou "Análise"
- Mostrar dezenas de categorias no mesmo gráfico
