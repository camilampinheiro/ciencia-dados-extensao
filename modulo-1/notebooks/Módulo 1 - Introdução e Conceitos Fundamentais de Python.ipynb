## 3. Variáveis e tipos de dados

Variáveis são espaços na memória usados para armazenar informações.

Em Python, alguns tipos básicos muito usados são:
- `str`: texto;
- `int`: números inteiros;
- `float`: números com casas decimais;
- `bool`: valores lógicos (`True` ou `False`).


````python
nome = "Ana"
idade = 20
altura = 1.65
estudante = True

print("Nome:", nome)
print("Idade:", idade)
print("Altura:", altura)
print("É estudante?", estudante)
````

### Verificando o tipo de cada variável

A função `type()` permite descobrir o tipo de dado armazenado em uma variável.


````python
print(type(nome))
print(type(idade))
print(type(altura))
print(type(estudante))
````

## 4. Operadores matemáticos

Operadores matemáticos são usados para fazer cálculos.
Os principais são:
- `+` adição
- `-` subtração
- `*` multiplicação
- `/` divisão
- `//` divisão inteira
- `%` resto da divisão
- `**` potência


````python
a = 10
b = 3

print("Soma:", a + b)
print("Subtração:", a - b)
print("Multiplicação:", a * b)
print("Divisão:", a / b)
print("Divisão inteira:", a // b)
print("Resto da divisão:", a % b)
print("Potência:", a ** b)
````

## 5. Operadores relacionais e lógicos

Operadores relacionais comparam valores:
- `>` maior que
- `<` menor que
- `==` igual a
- `!=` diferente de
- `>=` maior ou igual
- `<=` menor ou igual

Operadores lógicos combinam condições:
- `and` = e
- `or` = ou
- `not` = não


````python
x = 10
y = 5

print("x > y:", x > y)
print("x < y:", x < y)
print("x == y:", x == y)
print("x != y:", x != y)
print("x >= 10 and y < 10:", x >= 10 and y < 10)
print("x < 10 or y < 10:", x < 10 or y < 10)
print("not(x == y):", not(x == y))
````

## 6. Entrada e saída de dados

Esta é uma parte essencial para qualquer programa.

### Saída de dados
A **saída** acontece quando o programa mostra informações para o usuário.  
Em Python, isso é feito principalmente com `print()`.

### Entrada de dados
A **entrada** acontece quando o programa recebe informações do usuário.  
Em Python, isso é feito com `input()`.

> Observação: `input()` sempre retorna texto.  
> Quando quisermos números, precisamos converter com `int()` ou `float()`.


### Exemplo de saída de dados com `print()`


````python
mensagem = "Bem-vindo ao curso de Python!"
print(mensagem)
````

### Exemplo de entrada de dados com `input()`

> Esta célula pode pedir interação do usuário quando executada no Colab/Jupyter.


````python
nome_usuario = input("Digite seu nome: ")
print("Olá,", nome_usuario + "!")
````

### Convertendo a entrada para número

Quando o usuário digita um número, ele entra como texto.  
Por isso, fazemos a conversão antes de calcular.


````python
idade_usuario = int(input("Digite sua idade: "))
print("No próximo ano você terá", idade_usuario + 1, "anos.")
````

### Exemplo: média de dois números


````python
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))

media = (numero1 + numero2) / 2
print("A média é:", media)
````

## 7. Estruturas condicionais

As estruturas condicionais permitem que o programa tome decisões.

A estrutura mais comum é:

```python
if condição:
    # bloco de código
else:
    # outro bloco de código
```


````python
idade = 17

if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
````

### Exemplo com mais de uma condição


````python
idade = 20
tem_carteira = True

if idade >= 18 and tem_carteira:
    print("Pode dirigir")
else:
    print("Não pode dirigir")
````

## 8. Estruturas de repetição

Laços de repetição servem para executar uma ação várias vezes.

### `for`
Usamos `for` quando sabemos quantas repetições queremos fazer.

### `while`
Usamos `while` quando queremos repetir enquanto uma condição for verdadeira.


### Exemplo com `for`


````python
for i in range(5):
    print("Valor:", i)
````

### Exemplo com `while`


````python
contador = 0

while contador < 5:
    print("Contador:", contador)
    contador += 1
````

### Exemplo: números pares de 0 a 10


````python
for numero in range(0, 11, 2):
    print(numero)
````

## 9. Estruturas de dados básicas

Em Python, duas estruturas muito importantes para iniciantes são:
- **listas**: guardam vários valores em sequência;
- **dicionários**: guardam informações em pares de chave e valor.


### Listas


````python
frutas = ["maçã", "banana", "uva"]

print("Lista completa:", frutas)
print("Primeira fruta:", frutas[0])
print("Quantidade de frutas:", len(frutas))
````

### Percorrendo uma lista com `for`


````python
nomes = ["Ana", "Bruno", "Carla", "Daniel", "Eduarda"]

for nome in nomes:
    print(nome)
````

### Dicionários


````python
aluno = {
    "nome": "Carlos",
    "idade": 22,
    "curso": "Ciência da Computação"
}

print("Nome:", aluno["nome"])
print("Idade:", aluno["idade"])
print("Curso:", aluno["curso"])
````

## 10. Exemplo integrando entrada, processamento e saída

Todo programa costuma seguir esta lógica:
1. receber dados (**entrada**);
2. processar esses dados;
3. mostrar o resultado (**saída**).

Exemplo: cálculo da média de 3 notas.


````python
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))

media = (nota1 + nota2 + nota3) / 3

print("Média final:", media)

if media >= 6:
    print("Situação: aprovado")
else:
    print("Situação: reprovado")
````

## 11. Boas práticas iniciais

Escrever código não é apenas fazer funcionar.  
Também é importante que ele seja **claro, organizado e fácil de entender**.

### Evite nomes pouco descritivos


````python
x = 10
y = 8
z = (x + y) / 2
print(z)
````

### Prefira nomes mais claros


````python
nota1 = 10
nota2 = 8
media_notas = (nota1 + nota2) / 2

print(media_notas)
````

### Use comentários quando fizer sentido
Comentários ajudam a explicar a intenção do código, especialmente para iniciantes.


````python
# Calcula a média de duas notas
nota1 = 7
nota2 = 9
media = (nota1 + nota2) / 2

print("Média:", media)
````
