# Módulo 1 – Fundamentos de Python

Este notebook apresenta os conceitos introdutórios da linguagem Python, abordando sintaxe básica, variáveis, tipos de dados, operadores,
entrada e saída de dados, estruturas condicionais, estruturas de repetição, listas, tuplas, dicionários e boas práticas de programação.

```python
# Primeiro código
print("Olá, mundo!")
```

```python
# Variáveis
nome = "Ana"
idade = 19
altura = 1.65

print("Nome:", nome)
print("Idade:", idade)
print("Altura:", altura)
```

```python
# Tipos de dados
nome = "Camila"
idade = 20
nota = 9.0
curso = "Ciência da Computação"
aprovado = True

print(type(nome))
print(type(idade))
print(type(nota))
print(type(curso))
print(type(aprovado))
```

```python
# Operadores aritméticos
nota1 = 7.5
nota2 = 8.5

print("Soma:", nota1 + nota2)
print("Subtração:", nota1 - nota2)
print("Multiplicação:", nota1 * nota2)
print("Divisão:", nota1 / nota2)
```

```python
# Operadores relacionais e lógicos
nota1 = 7.0
nota2 = 9.0

print(nota1 > nota2)
print(nota1 == nota2)
print(nota1 != nota2)
print(nota1 > 5 and nota2 < 10)
```

```python
# Entrada e saída de dados
# nome = input("Digite seu nome: ")
# print("Olá,", nome)

nome = "Ricardo"
print("Olá,", nome)
```

```python
# Estrutura condicional
idade = 18

if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

```python
# Condicional com elif
nota = 7

if nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

```python
# Repetição com for
for i in range(5):
    print("Número:", i)
```

```python
# Repetição com while
contador = 0

while contador < 3:
    print("Contador:", contador)
    contador += 1
```

```python
# Listas
nomes = ["Ricardo", "Alvaro", "Camila"]

print(nomes)
print(nomes[0])

nomes.append("Bruno")
print(nomes)
```

```python
# Tuplas
cursos = ("Engenharia", "Direito", "Sistemas de Informação")

print(cursos)
print(cursos[1])
```

```python
# Dicionários
aluno = {
    "nome": "Carla",
    "idade": 20,
    "curso": "Engenharia"
}

print(aluno)
print(aluno["nome"])
print(aluno["idade"])
print(aluno["curso"])
```

```python
# Exemplo integrando conceitos
alunos = ["Ana", "Bruno", "Diego"]

for aluno in alunos:
    print("Aluno:", aluno)
```
