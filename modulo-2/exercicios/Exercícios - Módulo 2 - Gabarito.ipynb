# Módulo 2 — Gabarito — Lista de Exercícios

Este notebook contém o gabarito da lista revisada de exercícios com o dataset **Brazilian Cities**.
Ele tenta carregar o arquivo `brazilian_cities.csv`. Caso o arquivo não esteja disponível, usa uma pequena amostra didática para fins de demonstração.


```python
import pandas as pd

try:
    df = pd.read_csv("../datasets/brazilian_cities.csv")
    print("Dataset carregado com sucesso.")
except:
    print("Arquivo não encontrado. Usando amostra didática.")
    data = {
        "CITY": ["Fortaleza", "São Paulo", "Recife", "Salvador", "Curitiba", "Sobral", "Campinas"],
        "STATE": ["CE", "SP", "PE", "BA", "PR", "CE", "SP"],
        "STATE_REGION": ["NE", "SE", "NE", "NE", "S", "NE", "SE"],
        "GDP_CAPITA": [26000, 58000, 25000, 21000, 42000, 18000, 51000],
        "GDP": [70000000000, 700000000000, 40000000000, 60000000000, 80000000000, 4000000000, 65000000000],
        "IDHM": [0.754, 0.805, 0.772, 0.759, 0.823, 0.714, 0.805],
        "ESTIMATED_POP": [2700000, 12000000, 1600000, 2900000, 1900000, 210000, 1200000],
        "Motorcycles": [200000, 500000, 150000, 180000, 120000, 35000, 90000],
        "HOTELS": [50, 200, 40, 60, 70, 15, 80],
        "UBER": [1, 1, 1, 0, 1, 0, 1]
    }
    df = pd.DataFrame(data)

df.head()
```

## 1. Liste os 10 municípios com maior `GDP_CAPITA`

```python
df.sort_values(by="GDP_CAPITA", ascending=False).head(10)[["CITY", "STATE", "GDP_CAPITA"]]
```

## 2. Filtre apenas as capitais e compare `IDHM` e `GDP`

```python
df[df["CAPITAL"] == 1][["CITY", "STATE", "IDHM", "GDP"]]
```

## 3. Crie uma coluna de motos por habitante

```python
df["MOTOS_POR_HABITANTE"] = df["Motorcycles"] / df["ESTIMATED_POP"]
df[["CITY", "Motorcycles", "ESTIMATED_POP", "MOTOS_POR_HABITANTE"]].head()
```

## 4. Agrupe por `STATE` e descubra quantos municípios existem em cada estado no recorte analisado

```python
df.groupby("STATE")["CITY"].count().sort_values(ascending=False)
```

## 5. Agrupe os dados por estado `STATE` e calcule o PIB total (`GDP`) por estado

```python
df.groupby("STATE")["GDP"].sum().sort_values(ascending=False)
```

## 6. Verifique quais municípios têm `HOTELS` maior que zero e `UBER` igual a zero

```python
df[(df["HOTELS"] > 0) & (df["UBER"] == 0)][["CITY", "STATE", "HOTELS", "UBER"]]
```

## 7. Ordene os municípios por `ESTIMATED_POP` e observe os maiores valores

```python
df.sort_values(by="ESTIMATED_POP", ascending=False).head(10)[["CITY", "STATE", "ESTIMATED_POP"]]
```

## 8. Compare a média de `IDHM` por estado

```python
df.groupby("STATE")["IDHM"].mean().sort_values(ascending=False)
```

## 9. Calcule o PIB médio `GDP` por estado e ordene do maior para o menor

```python
df.groupby("STATE")["GDP"].mean().sort_values(ascending=False)
```

## 10. Exiba qual estado possui a maior média de `IDHM`

```python
df.groupby("STATE")["IDHM"].mean().sort_values(ascending=False).head(1)
```
