## Exercício 3 — Selecionando colunas

```python
df2 = df[[
    "CITY",
    "STATE",
    "CAPITAL",
    "ESTIMATED_POP",
    "AREA",
    "GDP",
    "GDP_CAPITA",
    "IDHM",
    "HOTELS"
]]
```

## Exercício 4 — Tratando valores nulos

```python
df2.isnull().sum()
```

```python
df2["HOTELS"] = df2["HOTELS"].fillna(0)
df2["IDHM"] = df2["IDHM"].fillna(df2["IDHM"].median())
```

```python
df2.isnull().sum()
```

## Exercício 5 — Removendo duplicatas

```python
df2.duplicated().sum()
```

```python
df2 = df2.drop_duplicates()
```

```python
df2.duplicated().sum()
```

## Exercício 8 — Criando novas colunas

```python
df["Cars"] = pd.to_numeric(df["Cars"], errors="coerce")
df["ESTIMATED_POP"] = pd.to_numeric(df["ESTIMATED_POP"], errors="coerce")
df["AREA"] = pd.to_numeric(df["AREA"], errors="coerce")
```

```python
df["DENSIDADE_DEMOGRAFICA"] = df["ESTIMATED_POP"] / df["AREA"]
df["CARROS_POR_HABITANTE"] = df["Cars"] / df["ESTIMATED_POP"]
```

```python
df[["CITY", "ESTIMATED_POP", "AREA", "DENSIDADE_DEMOGRAFICA", "Cars", "CARROS_POR_HABITANTE"]].head()
```
